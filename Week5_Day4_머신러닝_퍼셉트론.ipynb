```python
#퍼셉트론 만들기
class Perceptron:
    def __init__(self , lr = 0.1):
        self.w = 1
        self.b = 0 
        self.lr  = lr
        
    def fit(self , x , y): # x 는 입력 y 는 정답  p 는 모델의 출력으로 많이 쓸거임 
        p = self.predict(x)
        e = (y - p) / len(x) #얼마나 틀렸는지 
        self.w += self.lr * np.dot(e , x) # lr은 사람이 지정해주는 값 (leaning rate 학습률) 학습이 느리거나 빠르면 조절해줌. 
        self.b += self.lr * np.sum(e)     #이건 값 여러 개뽑을려고 함. 
        
    def predict(self , x):
        return x * self.w + self.b
```

```python
#퍼셉트론 만들기2 
class Perceptron:
    def __init__(self , lr = 0.1):
        self.w = 1
        self.b = 0 
        self.lr  = lr
        
    def fit(self , x , y): 
        p = self.predict(x)
        e = (y - p) / len(x)
        self.w += self.lr * np.sum(e + x) 
        self.b += self.lr * e
        
    def predict(self , x):
        return x * self.w + self.b
```

```python
import numpy as np

class Perceptron(object): 
    """Perceptron classifier.

    Parameters
    ------------
    eta : float
      Learning rate (between 0.0 and 1.0)
    n_iter : int
      Passes over the training dataset.
    random_state : int
      Random number generator seed for random weight
      initialization.

    Attributes
    -----------
    w_ : 1d-array
      Weights after fitting.
    errors_ : list
      Number of misclassifications (updates) in each epoch.

    """
    def __init__(self, eta=0.01, n_iter=50, shuffle=False, random_state=1):
        self.eta = eta
        self.n_iter = n_iter
        self.shuffle = shuffle
        self.random_state = random_state

    def fit(self, X, y):
        """Fit training data.

        Parameters
        ----------
        X : {array-like}, shape = [n_samples, n_features]
          Training vectors, where n_samples is the number of samples and
          n_features is the number of features.
        y : array-like, shape = [n_samples]
          Target values.

        Returns
        -------
        self : object

        """
        rgen = np.random.RandomState(self.random_state) #랜덤하게 생성. 
        self.w_ = rgen.normal(loc=0.0, scale=0.01, size=1 + X.shape[1]) 
        self.errors_ = []

        for _ in range(self.n_iter):
            errors = 0
            if self.shuffle:
                X, y = self._shuffle(X, y)
                
            for xi, target in zip(X, y): #xi 가 for문 150번돌고, 차례대로 self.predict에 들어감
                update = self.eta * (target - self.predict(xi)) # ★★★★★★★★ 이부분만 이해하면 퍼셉트론 이해잘한것. 
                #꽃하나를 예측을 하고 타겟을 비교를함.   update가 0 이면 맞춘것 
                self.w_[1:] += update * xi                      # ★★★★★★★★
                #w 는 x 만큼 곱해서 업데이트 해야하고 b 는 그대로 . (왜그런지는 월요일날 다시설명. )
                self.w_[0] += update                            # ★★★★★★★★
                errors += int(update != 0.0)
            self.errors_.append(errors)#에러몇개였는지 기록
            self.plot_decision_regions(X,y, self)#그림바뀌는것 확인하는것
        return self

    def _shuffle(self, X, y):
        """Shuffle training data"""
        r = np.random.permutation(len(y))
        return X[r], y[r]

    def net_input(self, X):
        """Calculate net input"""
        return np.dot(X, self.w_[1:]) + self.w_[0]

    def predict(self, X): # net_input -> X ㆎ W + b 
        """Return class label after unit step"""
        return np.where(self.net_input(X) >= 0.0, 1, -1) #이선보다 위에있으면 1 이라하고 선보다 아래에있으면 -1 이될거다. 
    #이때 setosa 는 -1 non setosa 는 +1 
        
    def plot_decision_regions(self, X, y, classifier, resolution=0.02):

        # setup marker generator and color map
        markers = ('s', 'x', 'o', '^', 'v')
        colors = ('red', 'blue', 'lightgreen', 'gray', 'cyan')
        cmap = ListedColormap(colors[:len(np.unique(y))])

        # plot the decision surface
        x1_min, x1_max = X[:, 0].min() - 1, X[:, 0].max() + 1
        x2_min, x2_max = X[:, 1].min() - 1, X[:, 1].max() + 1
        xx1, xx2 = np.meshgrid(np.arange(x1_min, x1_max, resolution),
                               np.arange(x2_min, x2_max, resolution))
        #print(xx1)
        #print(xx2)
        #print(np.array([xx1.ravel(), xx2.ravel()]).T)

        Z = classifier.predict(np.array([xx1.ravel(), xx2.ravel()]).T)
        Z = Z.reshape(xx1.shape)
        plt.contourf(xx1, xx2, Z, alpha=0.4, cmap=cmap)
        plt.xlim(xx1.min(), xx1.max())
        plt.ylim(xx2.min(), xx2.max())

        # plot class samples
        for idx, cl in enumerate(np.unique(y)):
            plt.scatter(x=X[y == cl, 0], y=X[y == cl, 1],
                        alpha=0.8, c=cmap(idx),
                        edgecolor='black',
                        marker=markers[idx], 
                        label=cl)
        plt.xlabel("sepal length cm")
        plt.ylabel("petal length cm")
        plt.legend()
        plt.show()
```
