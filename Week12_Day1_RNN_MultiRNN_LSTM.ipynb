```python
import numpy as np
size , seq_len = 100 , 3

X = np.empty(shape =(size, seq_len,1))
Y = np.empty(shape =(size,))

for i in range(size):
    c = np.linspace(i/10 , (i+seq_len-1)/10 , seq_len)
    X[i] = c[: , np.newaxis]
    Y[i] = (i + seq_len) / 10 

for i in range(len(X)):
    print(X[i] ,  Y[i]) 
```

```python
import tensorflow as tf
inputs = tf.keras.layers.Input(shape=(seq_len , 1))
x = tf.keras.layers.SimpleRNN(units = 20 , return_sequences= False)(inputs)
outputs = tf.keras.layers.Dense(1)(x)

model = tf.keras.Model(inputs= inputs , outputs= outputs )
model.compile(optimizer="adam", loss="mse")
model.summary()
```

```python
simple_rnn3 Param 개수  m(m+2) -> 20(20+2) -> 20x 22 -> 440
Dense Param 개수  20X1 + 1  -> 21 
```

```python
import matplotlib.pyplot as plt
history = model.fit(X , Y , epochs =300)
plt.plot(history.history["loss"])
plt.show()

y_hat = model.predict(X)
plt.scatter(Y,y_hat)
```

```python
import tensorflow as tf
inputs = tf.keras.layers.Input(shape=(seq_len , 1))
x = tf.keras.layers.SimpleRNN(units = 256 , return_sequences= False)(inputs)
outputs = tf.keras.layers.Dense(1)(x)

model256 = tf.keras.Model(inputs= inputs , outputs= outputs )
model256.compile(optimizer="adam", loss="mse")
model256.summary()
```

```python
import matplotlib.pyplot as plt
history = model256.fit(X , Y , epochs =300, verbose=0)
plt.plot(history.history["loss"])
plt.show()

y_hat = model256.predict(X)
plt.scatter(Y,y_hat)
```

```python
import tensorflow as tf
inputs = tf.keras.layers.Input(shape=(seq_len , 1))
x = tf.keras.layers.SimpleRNN(units = 34 , return_sequences= True)(inputs)
x = tf.keras.layers.SimpleRNN(units = 34 , return_sequences= True)(x)
x = tf.keras.layers.SimpleRNN(units = 34 , return_sequences= True)(x)
x = tf.keras.layers.SimpleRNN(units = 34 , return_sequences= False )(x)
outputs = tf.keras.layers.Dense(1)(x)

model_mul = tf.keras.Model(inputs= inputs , outputs= outputs )
model_mul.compile(optimizer="adam", loss="mse")
model_mul.summary()
 
history = model_mul.fit(X , Y , epochs =300, verbose=0)
plt.plot(history.history["loss"])
plt.show()

y_hat = model_mul.predict(X)
plt.scatter(Y,y_hat)
```

```python
장기 의존성 문제를 해결하기 위해 고안된 순환 신경망 셀이 LSTM(Long short term memory)

Cell states 상태를 하나더 넘겨줌 (memory) 그대신 역할을 바꿈. 
추가될수도 지워질수도있다., 셀상태가 모두 지워져 0이되면 출력을 내는일을 하지않게 된다 -> 

셀 상태를 지우는 망각게이트 (forget gate) 
= 앞서 만들어놨던 기억을 얼마나 유지할것인가 결정하는데 영향을 미침. -> 이전의 기억과 현재 입력 이 2개가 sigmoid에 작용해서 기억을 유지할것인가 영향미침

새로운 정보를 추가하는 입력게이트(input gate) 
= 이전출력과 현재입력이 똑같이 들어가는데 sigmoid와 tanh 에 동작을하고 이 두개가 곱셉을하고 기억의 합산이 됨.
tanh => RNN에 있던 tanh 와 같은역할, 현재기억에 입력데이터를 어떤형태로 얼마나 어떤모양으로 기억을 할 것 인가. 를 결정하는 얘 ->기억하기 용이한 vector로 바꿔줌
sigmoid => 기억하기 좋게 다듬어진 정보를 기억에 얼마나 강하게 인식시켜줄것인가. ex) sigmoid값이 0에가까우면 아무것도 추가적으로 기억하지않겠다는 뜻. 1이되면 tanh에 나온정보가 귀중하니 꼭기억각인시킴


셀 상태를 바탕으로 출력결정하는 출력게이트(output gate)
tanh 가있고 sigmoid 가 있음. 그런데 tanh에는 현재 입력까지 반영된 기억가지고 어떤정보를 출력할 것인가 를 결정하는게 tanh
sigmoid 이걸 얼마나 강하게 인식시켜줄것인가를 결정. 
```
