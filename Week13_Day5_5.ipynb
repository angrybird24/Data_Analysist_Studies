```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.ensemble import IsolationForest

setosa_df = data[data['species'] == 'setosa'][['sepal_length', 'sepal_width']]

X = setosa_df[['sepal_width', 'sepal_width']]

clf = IsolationForest(contamination=0.05, random_state=0)
outlier_labels = clf.fit_predict(X)


sns.scatterplot(x='sepal_length', y='sepal_width', data=setosa_df, hue=outlier_labels, palette={-1: 'red', 1: 'green'})
plt.title(' Outliers -> Red    ,    Normal -> Green')
plt.xlabel('Sepal Length (cm)')
plt.ylabel('Petal Length (cm)')
plt.legend()
plt.tight_layout()
plt.show()
```

```python
import numpy as np
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
from sklearn.datasets import load_iris
from sklearn.ensemble import IsolationForest

# 단계 1: iris 데이터셋 로드
data = load_iris()
iris_df = pd.DataFrame(data.data, columns=data.feature_names)

# 단계 2: Isolation Forest 적용
# 간단한 시각화를 위해 'sepal length (cm)'와 'petal length (cm)' 두 가지 feature를 사용합니다.
X = iris_df[['sepal length (cm)', 'petal length (cm)']]

# Isolation Forest를 이용하여 이상치 탐지
clf = IsolationForest(n_estimators=20, contamination=0.1, random_state=0)

# 단계 3: scatter plot 생성
plt.figure(figsize=(8, 6))
sns.scatterplot(x='sepal length (cm)', y='petal length (cm)', data=iris_df, hue=clf.fit_predict(X), palette={-1: 'red', 1: 'green'})
plt.show()
```
