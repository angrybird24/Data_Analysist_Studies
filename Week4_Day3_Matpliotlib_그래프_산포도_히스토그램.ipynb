```python
#그래프 관련 주요함수 

plt.style.use("ggplot") 

#한글 넣고싶을때 이러한 if 문 써줘야함. 
if os.name == 'posix':
    plt.rc("font", family="AppleGothic")
else:
    plt.rc("font", family="Malgun Gothic")
    
    
fig = plt.figure(figsize=(2, 2)) 
ax = fig.add_subplot(111)

data = [0,1]
ax.plot(data)
#그래프 그림 제목 설정 함수
ax.set_title("My 첫 Graph",fontdict={"fontsize":14,"fontweight":"bold"},loc="center")  #loc  -> center , left, right 

# X 축 Y 축 제목 설정 
ax.set_xlabel("X Axis",fontdict={"fontsize":14,"fontweight":"bold"},color="pink",loc="center")
ax.set_ylabel("Y Axis")

ax.set_xticks([1,3,5])
ax.set_yticks([10,20,30])

# X 축 Y 축 label 를 변경 
ax.set_xticklabels(['A','B','C'])
ax.set_yticklabels(['D',"E",'F'])


plt.show()
```

```python
import os
import pandas as pd
import matplotlib.pyplot as plt

if os.name == 'posix':
    plt.rc("font", family="AppleGothic")
else:
    plt.rc("font", family="Malgun Gothic")
    
base_url = ("https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/")
anime_stock_return_csv = os.path.join(base_url,"anime_stock_returns.csv")

df = pd.read_csv(anime_stock_return_csv)

df['Date'] =pd.to_datetime(df["Date"])
df = df.set_index("Date")

# df = pd.read_csv(anime_stock_return_csv, index_col=0)  이렇게 불러와서 바로 Date를 인덱스로 넣는방법도 있음 
df

#피켜 서브플롯 그래프 순으로 그리기 
plt.style.use('dark_background')
fig = plt.figure(figsize=(14,4)) 

ax = fig.add_subplot(111)
 
ax.plot(df.index, df["TOEI ANIMATION"], label = "KG ets 5 일봉")
ax.plot(df.index, df["IG Port"], label = "KG ets 15일봉")
#ax.plot 으로 만드는데 (df.index {인덱스안에 date있는걸 x 축으로} , 뒤에꺼는 TOEI ANIMATION 하고 IG Port 이 y 축으로 나와야함)

ax.set_title("KG ETS 5일선 15일선 주가",fontsize=20,fontweight="bold")
ax.set_xlabel("년월 ",fontsize=10)
ax.set_ylabel("주가 등락표",fontsize=10)

plt.legend(loc="upper right",fontsize=7) #legend는 plot 에서쓴 label을 만들어줌.  (loc 은 location의 약자로 상단 오른쪽으로 배치)
plt.show()
```

```python
import pandas as pd
import matplotlib.pyplot as plt

plt.style.use("Solarize_Light2")
url = "https://raw.githubusercontent.com/jbrownlee/Datasets/master/monthly-mean-temp.csv" 
df = pd.read_csv(url)
df = df.set_index("Month")

fig = plt.figure(figsize=(16,4))
ax1 = fig.add_subplot(111)
ax1.set_title("Temperature",fontsize=20 ,fontweight="bold",color="purple")
ax1.plot(df.index, df["Temperature"],label="Temperature" ,color='red') 

ax1.set_xlabel("Dates ",fontsize=10)
ax1.set_ylabel("Temp ",fontsize=10)

x_ticks = range(0,len(df.index),6) #눈금의 위치 
x_labels = [df.index[i] for i in x_ticks] # x축 또는 y 축 눈금 위치에 표시할 레이블을 지정하는 인자 
ax1.set_xticks(x_ticks,x_labels,rotation=60) #x 축 60도 꺾기   

ax1.legend()
plt.show()
```

```python
 
base_url = "https://raw.githubusercontent.com/practical-jupyter/sample-data/master/anime/"

anime_master_csv = os.path.join(base_url,"anime_master.csv")
df = pd.read_csv(anime_master_csv, index_col="anime_id")
df

#타잎종류 개수파악
types = df["type"].unique()
types

fig = plt.figure()
ax = fig.add_subplot(111)

for t in types:
    x = df.loc[df["type"] == t , "members"]  # members에서 type 이 t 인 경우 
    y = df.loc[df["type"] == t , "rating"]   
    ax.scatter(x , y , alpha=0.5, label = t) #alpha == 투명도 조절
    

plt.show()
```

```python
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt

iris = load_iris()
data = iris.data
target = iris.target
feature_names = iris.feature_names
 
fig = plt.figure(figsize=(16,8))
ax1 = fig.add_subplot(121)
ax2 = fig.add_subplot(122)

ax1.scatter(data[:,0] , data[:, 1] , c = target)
ax1.set_xlabel(feature_names[0])
ax1.set_ylabel(feature_names[1])
ax1.set_title("Sepal Length vs Width")

ax2.scatter(data[:,2] , data[:, 3] , c = target)
ax2.set_xlabel(feature_names[2])
ax2.set_ylabel(feature_names[3])
ax2.set_title("Sepal Length vs Width")

plt.show()
```

```python
import matplotlib.pyplot as plt

fig, ax = plt.subplots()

fruits = ['apple', 'blueberry', 'cherry', 'orange']
counts = [40, 100, 30, 55]
bar_labels = ['red', 'blue', '_red', 'orange']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:orange']

ax.bar(fruits, counts, label=bar_labels, color=bar_colors)

ax.set_ylabel('fruit supply')
ax.set_title('Fruit supply by kind and color')
ax.legend(title='Fruit color')

plt.show()
```
