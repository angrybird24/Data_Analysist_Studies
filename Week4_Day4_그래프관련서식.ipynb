```python
# 선의 색과 굵기 지정

x = np.array([0,1])
y = [0,1]


fig = plt.figure(figsize = (9,3))
ax = fig.add_subplot(111 , xlim =(-0.2,4)) 
ax.plot(x, y, linewidth = 1 , color='b' , label ="linewidth = 1(alpha= 1)") #linewidth 는 라인 굵기, color는 라인 색    
ax.plot(x + 0.6 , y , linewidth = 3 , color='g' , alpha=0.5,  label ="linewidth = 2(alpha=0.5)")  #x 에 더할수있는 이유가  ★★  
ax.plot(x + 1.2 , y , linewidth = 10 , color='r' , alpha=0.7, label ="linewidth = 10 (alpha=0.7)")# x는 np이기때문 배열 y는 안됨! 
plt.legend()
plt.show()
```

```python
#선의 끝 모양 지정 
x = [0.1, 1]
y = np.array([0, 0])
 
fig = plt.figure(figsize = (6,3))
ax = fig.add_subplot(111 , xlim =(0 , 1.7) , ylim=(-0.2 , 1.2))    
ax.plot(x  , y+ 0.6 , linewidth = 12, solid_capstyle='butt' , color='crimson' , label ="butt")  
ax.plot(x  , y+ 0.2 , linewidth = 12, solid_capstyle='round' , color='lightgray' , label ="round")  
ax.plot(x  , y+ 0.9 , linewidth = 12, solid_capstyle='projecting' , color='cyan' , label ="projecting")    
                                     #solid_capstyle 은 선의 끝부분. 'butt' (기본값): 선이 끝날 때 직선
                                     #                               'round': 선이 끝날 때 둥근 모양 
                                     #                               'projecting': 선이 끝날 때 선의 방향으로 약간 돌출

plt.legend(loc=4) #loc 1~4 로 모서리쪽에 배치가능. 
plt.show()
```

```python
#선의 결합점 모양 지정  

y = [0, 1.0, 0.2, 1.2]
x = np.array([1, 2, 3, 4])  

fig = plt.figure(figsize=(6, 3))
ax = fig.add_subplot(111, xlim=(0, 4.5))

ax.plot(x, y, linewidth=2, solid_joinstyle="miter", label="Miter" )
ax.plot(x + 0.1, y, linewidth=2, solid_joinstyle="round", label="Round" )
ax.plot(x + 0.2, y, linewidth=2, solid_joinstyle="bevel", label="Bevel" )

ax.legend()
plt.show()
```

```python
#선의 종류 지정
x = np.array([0,1])
y = np.array([1.9,1.9])
#linestyle = 'solid' == '-' , 'dashed' == '--' , 'dotted' == ':' , 
#선종류 변경 

fig = plt.figure(figsize=(6, 3))
ax = fig.add_subplot(111, xlim = (-0.1 , 1.7) , ylim = (0,2))

ax.plot(x , y, linewidth=5, linestyle = 'solid', solid_joinstyle="miter", solid_capstyle='butt', color='blue', label="solid")
ax.plot(x , y-0.1, linewidth=5, linestyle = '-', solid_joinstyle="miter", solid_capstyle='butt', color='blue', label="-")

ax.plot(x , y-0.4, linewidth=5, linestyle = 'dashed', solid_joinstyle="miter", solid_capstyle='butt', color='crimson', label="dashed")
ax.plot(x , y-0.6, linewidth=5, linestyle = '--', solid_joinstyle="miter", solid_capstyle='butt', color='crimson', label="--")


ax.plot(x , y-0.8, linewidth=5, linestyle = 'dotted', solid_joinstyle="miter", solid_capstyle='butt', color='cyan', label="dotted")
ax.plot(x , y-1, linewidth=5, linestyle = ':', solid_joinstyle="miter", solid_capstyle='butt', color='cyan', label=":")


plt.legend()
plt.show()
```

```python
#산점도(산포도) 에서 점의 색과 부루명도 지정 

x = np.arange(0, 6 , 0.1)
y = np.sin(x)
z = np.cos(x)

fig = plt.figure()
ax = fig.add_subplot(111)

ax.scatter(x , y , label= "sin", color = "lightblue", edgecolor = "blue" )

ax.scatter(x , z , label= "cos", color = "lightblue", edgecolor = "blue", alpha =0.3 )

plt.legend(framealpha=0.5 , loc='lower right') #framealpha 하면 label에 있는 부분이 투명해짐. 
plt.show()
```

```python
# plt.boxplot()
from sklearn.datasets import load_iris 
import matplotlib.pyplot as plt
import pandas as pd

plt.style.use("classic")  
iris = load_iris()
X = iris["data"] 
# iris.keys()
df = pd.DataFrame(X , columns = iris.feature_names)
df

colors = ['lightblue', 'lightgreen', 'lightpink',"red"]

plt.figure()
# 박스 그래프의 색상 지정 
boxplot = plt.boxplot(df, labels=df.columns, notch=True, patch_artist=True) #patch_artist=True를 설정해야 박스 그래프를 색칠가능

for patch, color in zip(boxplot['boxes'], colors): #각각의 박스에 색깔 4개 넣어주기 
    patch.set_facecolor(color)
    
    
plt.xticks(rotation=15, color="blue")

    
plt.show() #빨간선이 중앙값 -> 50% 위치한값 |  제일윗,아랫 iqr 정상,최하범위  | 바그래프 젤위 : 상위 20%  젤아래: 하위20%
```

```python
#수평선 , 수직선 긋기 
r = np.random.rand(10)
plt.figure(figsize=(5,2))
plt.plot(r)
plt.axhline(0.75, color="red") 
plt.axvline(6 , color="green")
plt.show()
```

```python
#수평선 axhline , 수직선 긋기 axvline 2 
r = np.random.rand(10)
plt.figure(figsize=(5,2))
plt.plot(r)
plt.axhline(0.75, xmin = 0.1, xmax = 0.5, color="red")  # A ~ B 길이까지 설정 
plt.axvline(6 ,  ymin = 0.2, ymax = 0.7, color="green") 


plt.show()
```

```python
# 부분 색칠 axvspan()
r = np.random.rand(10)
plt.figure(figsize=(5,2))
plt.plot(r)
plt.axhline(0.75, color="red") 
plt.axvline(6 , color="green")
plt.axhspan(0.25 , 0.5 , alpha=0.3 , color='orange', xmin=0.6, xmax=0.8) #2 ~ 4 까지 색칠됨 , Alpha 는 필수 !  #여기도 범위지정가능
plt.axvspan(2,4 , alpha=0.3)  #2 ~ 4 까지 색칠됨 , Alpha 는 필수 ! 
plt.show()
```

```python
# hatch , fill 채우는공간 또 꾸미기 
r = np.random.rand(10)
plt.figure(figsize=(5,2))
plt.plot(r)
plt.axhline(0.75, color="red") 
plt.axvline(6 , color="green")
plt.axhspan(0.25 , 0.5 , alpha=0.3 , color='orange', xmin=0.6, xmax=0.8, hatch="*", fill=False) 
plt.axvspan(2,4 , alpha=0.3)                                  # hatch로 색칠부분 표시가능 # / *   #fill 로 색 빈칸으로만하기
plt.show()
```

```python
#fill_between() 영역만큼 색칠 
r1 = np.random.rand(10)
r2 = np.random.rand(10)

 
plt.figure(figsize=(5,2))
plt.plot(r1)
plt.plot(r2)
plt.axhline(0.75, color="red") 
plt.axvline(6 , color="green")
plt.axhspan(0.25 , 0.5 , alpha=0.3 , color='orange', xmin=0.6, xmax=0.8, hatch="*", fill=False) 
plt.axvspan(2,4 , alpha=0.3) 
plt.fill_between(np.arange(2,7) , r1[2:7] )
plt.show()
```

```python
#fill_between() 영역만큼 색칠 2  A 라인과 B 라인사이를 색칠 
#두그래프 사이의 차이를 얘기할때 이걸 많이씀 ! 
#ex) 물쓰는 수도데이터라 했을때 1데이터는 물을 보낸데이터고 2데이터는 물을쓰는 그래프 일때 2개랑비교해볼때
#    같아야 하는데 그 차이를 볼때 이걸 쓸수있음. or 어디누가 물을 빼내거나 어디서 손실이 있는걸 파악 할수있음.! 

r1 = np.random.rand(10)
r2 = np.random.rand(10)
 
plt.figure(figsize=(5,2))
plt.plot(r1)
plt.plot(r2)
plt.axhline(0.75, color="red") 
plt.axvline(6 , color="green")
plt.axhspan(0.25 , 0.5 , alpha=0.3 , color='orange', xmin=0.6, xmax=0.8, hatch="*", fill=False) 
plt.axvspan(2,4 , alpha=0.3) 
plt.fill_between(np.arange(2,7) , r1[2:7] , r2[2:7],color="red", alpha=0.4)
plt.show()
```
