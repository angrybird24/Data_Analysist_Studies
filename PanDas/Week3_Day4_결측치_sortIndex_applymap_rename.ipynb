```python
df = pd.DataFrame([[0.3,0.2,0.6,0.9],[0.7,0.3,0.3,0.6],
                   [0.1,0.5,0.7,0.5],[0.6,0.4,0.1,0.3],
                   [0.8,0.9,0.2,0.4]],
                 columns=['A','B','C','D'])
#loc 사용 c 열 2 행 출력
print(df.loc[2,"C"],"\n") 

# #iloc 사용 2열 3행 출력
print(df.iloc[2][3],"\n")

# #loc 사용 B~D  1~3 행 출력
print(df.loc[1:3,"B":"D"])

# #iloc 사용 0~2열 의 2~4행 출력
print(df.iloc[2:5,0:2])
```

```python
iris = load_iris() 
iris_df = pd.DataFrame(data=iris.data , columns = iris.feature_names)
iris_df['target'] = iris.target #target 열 추가 
iris_df
```

```python
#sepan length (cm) 열의ㅡ 값 모두 선택
iris_df.loc[:,"sepal length (cm)"]

#인덱스가 2인 행 출력 (iloc 사용)
iris_df.iloc[2]

#인덱스 라벨이 2인 행을 출력 (loc 사용)
iris_df.loc[2]

#인덱스가 3인 행부터 7인 행까지 선택 (iloc사용)
iris_df.iloc[3:8]

#인덱스 라벨이 3 행 ~ 7행까지 선택 (loc 사용)
iris_df.loc[3:7]

#인덱스 라벨이 5인 행의 petal length (cm)  열값을 출력
iris_df.loc[5]["petal length (cm)"]

#전체 데프에서 	sepal width (cm) 열에서 3보다 큰 행만 선택  
iris_df[iris_df['sepal width (cm)'] >3]
```
