```python
#라이브러리 호출
import pandas as pd
import numpy as np
from datetime import datetime 
```

```python
#CSV파일 불러오기
df = pd.read_csv('C:/Users/carlo/Desktop/AI_CWNU/documents/COVID19.csv', encoding="cp949")
df
```

```python
#위에 카테고리들 한글에서 영어로 바꿔주기 
df.columns=['Category','Type','Date','Headcount','Unnamed'] #ex) 구분별 상태별 시점 -->>Category','Type','Date' 
df

#위에 값중
#strtype 인 Date 값들을 Timestamp type 으로 바꿈 
df['Date'] =pd.to_datetime(df["Date"])
df

#Date 에 있는걸 index로 변경 
df = df.set_index("Date")
df
```

```python
#Unnamed 없애기 (컬럼삭제)
df.drop("Unnamed",axis=1,inplace=True)
df
```

```python
#결측치 개수 컬럼별 확인 
df.isna().sum() 
df.isnull().sum()

#이렇게하면 NaN 인 것들만 불러냄 
print(len(df[df.isna().any(axis=1)]))
df[df.isna().any(axis=1)]
```

```python
#결측치가 존재하는 행 -> 결측치가 있는 열만 지우고 남은 값들만 출력
df = df.dropna(axis=0)
df
```

```python
#문제 1. 총확진자 수는 몇명인가.? 
df["Headcount"].sum()

#문제 2. "Category" 별, 'Type' 별 인원수 확인 ? 
g = df.groupby("Category")
a=g["Headcount"].sum()
a

#문제 3  년도 별로 데이터 프레임을 분리 
s = df.sort_index() #일별 정리 먼저 해주기  (그전에 인덱스위치에 stamp type으로 변경된 date를 먼저 넣어야함)
y = s.resample("Y").sum() #그리고나서 resampling 해주는데 . Y -> Year 별로 sum -> 더한 모든값. 
#sum() 안에 numeric_only=True 이거 넣어주면 빨간색 없어짐. 
y # 이때 모든 년도 더한값이 df["Headcount"].sum() 이것이랑 같아야함! 

```

```python
# "통계확인"
df.describe()


#피봇 테이블
pivot_table = df.pivot_table(index = "Category",columns = "Type", values = "Headcount", aggfunc = "max")
pivot_table

#교차 일람표 
cross_tab = pd.crosstab(index = df["Category"], columns=df["Headcount"])
cross_tab

# "Headcount" 열에 존재하는 음의 값을 제거 
df = df.loc[df["Headcount"]>=0]
df

#제거후 다시하면 -값 다 사라져있음
cross_tab
```

```python
#다운샘플링
downsampled_df = df.resample("M").sum(numeric_only=True)
downsampled_df
```

```python
#plot()   데이터 프레임 인덱스를 x 열 데이터를 y 축으로 그래프 그려줌 
downsampled_df.plot()
```

```python
#롤링 윈도우 (Rolling Window)
rolling_mean = downsampled_df["Headcount"].rolling(window = 5).sum()
rolling_mean
# 롤링 윈도우는 데이터 분석에서 데이터의 특성을 파악하고 변화를 추적하는 데 유용한 도구
# 데이터의 패턴 파악, 이상치 탐지, 추세 분석 등 다양한 분석 작업에 적용될 수 있으며,
# 데이터 처리 및 모델링 과정에서 중요한 역할을 합니다.
```

```python
#★★★★★★★★★실습! ★★★★★★★★★★ 할거임. 
tif = pd.read_csv('C:/Users/carlo/Desktop/AI_CWNU/documents/TIF.csv', encoding="cp949")

#1. Measuredate   ,  MeasureType , Value  이렇게 3개 살리기 근데    Measuredate 형변환후 -> index 이렇게 
#2  피봇테이블 만들기 -> date -> index type 이 열
#3 요걸로 시간에 따른 판형을 -> moving average 해서 (데이터가 1초마다들어옴) 윈도우 를 1분 단위로 . 
#4 어느 센서가 상관도가 높은지 top 3 뽑기 
tif.drop(tif.loc[:, "DETAILVALUE1":"EDITDATE"].columns, axis=1, inplace=True) 
tif.drop(tif.loc[:, "QUALITY":"DATATYPE"].columns, axis=1, inplace=True) 
tif.drop(columns="MACADDRESS",inplace=True) 
tif.drop(columns="IFSEQ",inplace=True) 
tif
```

```python
tif['MEASUREDATE'] =pd.to_datetime(tif["MEASUREDATE"])
tif = tif.set_index("MEASUREDATE")
tif
```

```python
pivot_table = tif.pivot_table(index = "MEASUREDATE",columns = "MEASURETYPE", values = "VALUE" )
pivot_table
```

```python
columns_to_average = ['humi', 'vibr', 'powe', 'temp', 'tira', 'nois']
rolling_mean = pivot_table.loc[:, columns_to_average].rolling(window='60s').mean()
rolling_mean
```
