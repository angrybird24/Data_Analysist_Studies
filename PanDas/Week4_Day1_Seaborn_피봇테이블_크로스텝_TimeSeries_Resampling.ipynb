```python
# 피봇 테이블 pivot_table
# 데이터를 요약하고 재구성하기 위한 기능을 제공하는 데이터 분석 도구 
df = sns.load_dataset("titanic")

pivot_table = df.pivot_table(index="sex",columns="class",values="survived",aggfunc='mean') #aggfunc 집계 함수 aggregation function
pivot_table
```

```python
#롤링 윈도우 (Rolling Window)
data = {"date":pd.date_range(start="2023-05-01",end="2023-05-10", freq="D"), 
        "value":[1,6,4,12,36,17,44,91,180,23]} 
df = pd.DataFrame(data)

df.set_index("date",inplace=True)
df
```

```python
rolling_mean = df["value"].rolling(window=5).mean()
rolling_mean

#시리얼 분석할때 필요함 window 5개 되있으면 앞에 5개를 더하고 5개를 나눈걸 5번째, 6번째는 6개 더하거 6나눈걸 6번째 1~4 는없음! 

[print(i) for i in rolling_mean]
```
