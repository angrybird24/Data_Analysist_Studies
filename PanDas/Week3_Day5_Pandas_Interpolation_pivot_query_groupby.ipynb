```python
#사용자 정의 집계 함수 정의 
iris = load_iris() 
df = pd.DataFrame(data=iris.data , columns = iris.feature_names)

def custom_summary(x):
    summary = {"mean":x.mean(),
              "median":x.median(),
              "std":x.std(),
              "max":x.max(),
              "min":x.min()
              }
    return pd.Series(summary)

result = df.apply(custom_summary)
result
```

```python
data ={"City":["Seoul","Seoul","Busan","Busan","Incheon"],
      "Population":[9.7,9.8,3.4,3.6,2.0],
      "GDP":[620,650,180,200,150],
      "Area":[605.2 , 605.2 , 770.1 , 770.1 , 1049]}
df = pd.DataFrame(data)
grouped = df.groupby("City")
for i in grouped:
    print(i)
    
def max_min(column):
    return column.max()-column.min()

#중복된 값을 MAX- MIN 뺀값  // 인천이 0 인 이유는 비교대상이 없어서 
result = grouped[["Population","GDP"]].apply(max_min)
result
```

```python
# aggregate(집계)
data ={"City":["Seoul","Seoul","Busan","Busan","Incheon"],
      "Population":[9.7,9.8,3.4,3.6,2.0],
      "GDP":[620,650,180,200,150],
      "Area":[605.2 , 605.2 , 770.1 , 770.1 , 1049]}
df = pd.DataFrame(data)

grouped = df.groupby("City") 
agg_result = grouped.agg({"Population":"sum","GDP":["mean","max"],"Area":"min"})
agg_result
```
