```python
시계열 분석이란?
시간에 따라 관측된 자료(data)를 시계열 자료 라고함.
(시간에 대해서 어떤 특정한 값을 갓는 데이터의 나열. )

시계열 구성 요인 4 가지 
1. 추세 요인(Trend Factor) : 데이터의 전체적인 방향성  
    (트렌드는 시계열 데이터가 오랜 기간 동안 어떻게 증가하거나 감소하는지를 나타냄)

2. 계절성 요인(Seasonality Factor) : 절성은 시계열 데이터가 일정한 주기로 반복되는 패턴을 보이는 것을 의미
    
    
3. 순환성 요인 (Cyclicity Factor) : 일정하지 않은 주기로 반복되는 패턴
    계절성과 유사하지만 주기가 정확하게 일정하지 않을 수 있음 ex) 주식의 차트 up and down 
    
    
4. 불규칙한 편동성 요인(irregular fluctuations ): 시계열 데이터에 무작위한 변동이나 오차를 추가하는 요소
```

```python
가법 모형 Y(t) = T(t) + S(t) + C(t) + ε(t) 
# 서로 독립적 -  트렌드, 계절성, 순환성, 잡음 등의 요소들이 독립적으로 더해져서 시계열을 형성

승법 모형 Y(t) = T(t) × S(t) × C(t) × ε(t)
#서로 상호 의존적 - 트렌드, 계절성, 순환성, 잡음 등의 요소들이 상호간에 곱해져서 시계열을 형성

Y(t) = T(t) × S(t) × C(t) × ε(t)
Y(t): 시간 t에서의 관측값
T(t): 트렌드 요소
S(t): 계절성 요소
C(t): 순환성 요소
ε(t): 잡음 요소
    
    
```

```python
백색 잡음(White Noise)은 시계열 데이터에서 가장 간단한 형태의 잡음
  이러한 잡음은 예측할 수 없는 무작위한 변동을 나타내며 어떠한 패턴도 가지지 않음

    
확률 보행(Random Walk)은 시간에 따라 누적된 무작위한 변동을 나타내는 시계열 데이터
  확률 보행은 이전 관측값에 현재 시점에서의 백색 잡음을 더한 값으로 이루어짐
    (랜덤으로 움직이는데 이전값을 기준으로 함 )
    
```

```python
정상성(Stationary) (시계열이 얼마나 일정하게 반복되는가.) (정상성을 뛸수록 미래를 예측하기 쉬움 )
다음 세 조건을 만족하는 확률과정을 정상시계열 이라고함.

1 평균이 일정 #트랜드가 변화지않아야한다 ex)기울기가 0으로 쭉 가야한다.  (그래프의 기울지 않은 한 쪽으로만 가는 방향)

2 분산이 존재하며 상수(constant) # 분산이 커지거나 작아지는게 아니라 일정해야한다.  (그래프의 진폭)

3 두 시점 사이의 자기 공분산은 시차에만 의존 # (패턴의 모양이 바뀌지 않아야한다.)
```

```python
Autoregressive 모델(AR 모델) 이전 시간대의 관측값들에 대한 가중합으로 현재 시간대의 값을 예측하는 선형 모델
hyperparameter : p 

Moving Average 모델 (MA 모델) MA 모델은 이전 시간대의 잔차값들에 대한 가중합으로 현재 시간대의 값을 예측하는 선형 모델
hyperparameter : q 
(오차들이 모여서 만들어진 관점) ex) 고요속의 외침 게임같은거 귀마개쓰고 단어 맞추기 뒤로전달전달해서 

T 시점 엡실론 = y 시점 엡실론 - 1/q(평균) * (과거의 y 의 평균)
                                 Error  =   그추세와 현재 추세의 차이 
    
Y(t) = c + Σ(θ(i) * ε(t-i)) + ε(t)

여기서,

Y(t): 현재 시간대(t)의 관측값
c: 상수항 (잔차의 평균)
θ(i): i번째 이전 시간대의 가중치 (MA 계수)
ε(t-i): i번째 이전 시간대의 잔차값
ε(t): 잔차 (오차항)
    
    
Autoregressive Moving Average 모델(ARMA)  위에 2개의 식을 합친 것 ( AR(p) 모델과 MA(q) 모델을 결합한 형태 )
 과거랑 현재를 둘다 예측 -> 이전 시간대의 관측값과 잔차값들에 대한 가중합으로 현재 시간대의 값을 예측하는 선형 모델
    
hyperparameter :p,q 
    
    
ARIMA(Autoregressive Integrated Moving Average) 모델은 AR, MA, 및 I 성분을 결합한 시계열 예측 모델
     ARIMA 모델은 시계열 데이터의 추세(Trend)와 계절성(Seasonality)을 모델링하고 예측하는 데 사용
    
차분(Difference, I) 성분: d
현시점 데이터에서 d 시점 이전 데이터를 뺀 것. 
차분 성분은 시계열 데이터의 추세를 제거하기 위해 현재 시간대의 관측값과 이전 시간대의 관측값의 차이를 계산함 
(미분 )
```

```python
p ,q ,d 를 찾을 떄 ACF PACF 가 필요함. 

ACF (Autocorrelation Function): 자기 상관
    ACF는 시차(lag)에 따른 관측값들 사이의 자기상관성을 측정하는 함수
    k 를 0부터 T 까지 바꿔가면서 자기상관값이 바뀌어가는냐 바로 직전 과거의 값을 상관이 높을것
    그치만 태초로 가면갈수록 그게 옅어지면서 상관계수는 작아질것임. 
 

PACF (Partial Autocorrelation Function): 부분 상관
    PACF는 다른 시차(lag)들의 영향을 제외한 두 관측값 간의 자기상관성을 측정하는 함수
    
    
AR 인지 MA 인지 확인하는법
```

```python
import numpy as np
import pandas as pd
import statsmodels as sm
# import statsmodels.api as sm

import matplotlib.pyplot as plt
```

```python
data = sm.api.datasets.co2.load_pandas()
data
```

```python
y = data.data
y.info()
```

```python
plt.figure(figsize= (15,5))
plt.plot(y["co2"])
plt.show()
#결측치 때문에 중간중간에 끊킨부분이 있음.
```

```python
#결측치 보완
tmp = y["co2"].resample("M").mean() 
tmp.plot() #이래도 안되는 이유가 resampling 은 W 기간동안 그사이에 값들을 평균을 내주는데 그값이 없으면 평균을 못내기 때문. 
#이때는 M 보단 2M 로  3M 로 늘려서 없앨수있음 또는 

# 뒤에 있는걸 앞으로 채우기 채우기
y = tmp.bfill()
y.plot()
```

```python
y
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(y , model = "addictive")#, freq = )
result.plot()
```

```python
from statsmodels.graphics.tsaplots import plot_acf ,plot_pacf

plot_acf(y)
plot_pacf(y)

plt.show()
```

```python
p <- pcaf 가 cut - off 일어나기 직전 2 시차 
q <- acf 가 점진적으로 감소함으로 0 
```

```python
#차분은 데이터가 정상상태가 될때까지
y_diff = y.diff()[1:] #period = n -> n 차 차분 

plot_acf(y_diff)
plot_pacf(y_diff)

plt.show()
```

```python
y_diff.plot()
```

```python
y_diff = y.diff(periods = 2 )[2:] 
plot_acf(y_diff)
plot_pacf(y_diff)

plt.show()
```

```python
y_diff.plot()
```

```python
d <- 1차분으로 정상화됨
p , d, q = 2, 1, 0 
```

```python
from statsmodels.tsa.arima.model import ARIMA

arima = ARIMA(y , order = (2,1,0))
arima_result = arima.fit()
```

```python
arima_result.summary()
```

```python
AIC , BIC ,HQIC 를 비교해서 더 작은게 더 좋은거임. 
ARIMA(2,1,0) 이 파라미타를 바꿔가면서 변경함.  그걸 비교 


```

```python
p = arima_result.predict()
plt.figure(figsize = (10, 5))
plt.plot(y )
plt.plot(p[1:])
plt.show()
```

```python
arima_result.plot_diagnostics()
```

```python
f =  arima_result.forecast(steps = 120)
f
```

```python
plt.figure(figsize= (15,5))
plt.plot(y)
plt.plot(f)
plt.show()
```

```python
fore = arima_result.get_forecast(steps = 120)
fore
```

```python
plt.figure(figsize= (15,5))
plt.plot(y)
plt.plot(fore.predicted_mean)

ci = fore.conf_int()
plt.fill_between(ci.index , ci.iloc[:,0], ci.iloc[:,1],
                color="black", alpha=0.2)

plt.show()
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX
arima = SARIMAX(y , order = (1,1,1),
               seasonal_order=(1,1,1,12),
               enforce_stationarity=False,
               enforce_invertibility=False)
arima_result = arima.fit()
fore = arima_result.get_forecast(steps=600)

plt.figure(figsize=(10,5))
plt.plot(y)
plt.plot(fore.predicted_mean)
ci = fore.conf_int()

plt.fill_between(ci.index, ci.iloc[:,0] , ci.iloc[: ,1] , color = "black" , alpha = 0.2)
plt.show()
```
