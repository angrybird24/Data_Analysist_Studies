```python
이상탐지(Anomaly) 정의 및 분석
이상탐지는 현업에서 이상이라고 정의한 사건 및 패턴을 탐지하는 활동 

기업에서 이상탐지목적
제조 심각한 고장발생 전 이상탐지를 통해 심각한 RISK 방지
금융 비정상적 거래 및 사기 거래 방지를 통해 소비자 보호
게임 버그 유저 및 비정상 유저 탐지를 통해 게임 정상운영

이상치와 이상은 틀림.
이상치(Outlier) 은 범위에서 벗어낫다
 데이터의 관점
 관된 데이터의 범위에서 많이 벗어난 (거리) 값
 분석하고자 하는 데이터에서 매우 적은 확률로 나타나는 데이터
 분석결과 해석시 문제를 발생시켜 사전에 제거 되어야할 대상(전처리과정)

이상(Abnormal) 은 Threshhold 가있다 (휴대폰온도 100도 이상)
 문제해결의 관점
 현업의 Domain에서 볼때 문제 발생가능성이 높은 데이터
 ★정상적인 범주에 있는 데이터라도 이상이라고 볼수있음-> 고장이나 불량 야기, 이상치데이터일수도있음 
 관리하지않으면 큰 위험이 있을 것 같은 데이터에 적용되어야 하며 , 일반적으로 자주 발생되진 않음
```

```python
이상 데이터 발생 원인
1 표본추출 오류 (Sampling Error)
    샘플링 분석을 통해 모수를 예측하는데(시간과 돈을 절약하기위해) , 다른 모수(전체 (모든수))에서 추출하는 경우 ,즉
    이상치 데이터가 포함된 샘플링을 한경우 , 학습데이터와 테스트 데이터간의 성능차이가 큰경우(균등한 데이터 추출이 안된경우)

2 입력오류 (Data Entry Error)
    데이터수집, 기록하는 과정에서 발생하는 Human Error , 데이터 분포로 쉽게 탐지가능, 임상,
    의료데이터는 아직 까지 사람이 입력하고있음

3 실험오류 (Experimental Error)
    실험을 통해 데이터를 수집하는경우 ,  실험조건이 동일하지 않은 경우 , 서로 다른 조건의
    챔버를 사용해서 데이터(온습도, 미세먼지) 를 수집하는 경우 
(ex: 가장맛있는 라면을 실험을할때 -> 라면을 끓이는데 생수 , 물의 양 이 동일하고, 끓는물 100도일때, 끓는물 95 도일때 (나머지는같음) 근데 알고보니 물의양이 틀림. -> 실험오류) 

4 측정오류 (Measurement Error)
    측정장비를 통해 데이터를 수집한경우 측정기 자체에 오류(노후화로 인한)

5 데이터 처리 오류 (Data Preprocessing Error)
    데이터(ETL (Extract > Transform > Load) 과정에서 발생되는 데이터 처리오류 , 가장 많이발생됨(데이터 엔지니어가
    데이터 분석가에게 전달하는 과정에서 발생되기도함 , 서버 셧다운으로 인한 경우)

6 자연오류 (Natural Outlier), 진짜 이상 데이터
    자연스럽게 발생되는 이상값, 우리가 데이터분석을 통해 해결하려고 하는 문제 
```

```python
이상 탐지 종류
┌ Time Series (Sequential) -> 주식 
└ Static (정적인, Point data) -> 키 , 몸무게 , 집값


┌ Univariate(단변량) -> 한사람을 설명하는 데이터가 1개 일때 사람이름 있고 키있고 2개만
└ Multivariate(다변량) ->  한사람을 설명하는 데이터가 여러개 일때  사람이름있고 키있고 무게있고 주소있고 ...

Data Type(binary/ categorical / continuous / hybrid)
            1/0  / 10,20,30대  / 173.2cm(정수나 소수점있는것) / hybrid는 섞인것 
(Discrete -> boolean 처럼 1/0 또는 10,20,30  임 , Univariate,Multivariate 안에 Discrete,  Continuous 이렇게 2가지로 나뉨 밑은 또다른게 나눠짐)

┌Relational (상관관계있는)  선형회귀 사용시 상관관계가 높은 변수는 버린다(다중공선성 때문에)
└Independent(독립적인) 


┌ Well Known  (기존 룰 적용가능) 잘알려진 문제는 많은 자료가 있다. (예 : 설비 센서 이상데이터 탐지) , 현업의 엔지니어 경험도 가지고와야한다(로직화)
└ or Not  (잘 알려저 있지않은) 알려져있지 않은 문제는 다양한 알고리즘 적용이 필요함


※5번부터 진행하여 현업의 엔지니어와 소통, 기존자료확인 
```

```python
이상 탐지의 종류 (문제 해결 유형)
1 Point Anomaly Dectection (포인트 이상탐지)
    측정된 데이터에서 정적인 점 분포에 초점
    특정 Point 의 이상치를 감지
    일반적으로 말하는 데이터 내 Outlier 를 거리기반으로 dection


2 Contextual Anomaly Detection (상황적 이상 탐지)
    시계열과 같은 동적인 특성에 초점(과거에 데이터가 현제 값에 영향을 미칠때, sequential)
    연속적인 변화 패턴을 읽어 이상치 감지 , 맥락을 고려해서 예상변활와 동떨어진 결과를 감지


3 Collective Anomaly Detection (집단적 이상 탐지)
    개별 인스턴스가 아닌 집합 인스턴스 비교를 통해 Local 한 이상치를 확인 (집합으로 봤을때는 정상, 개별로 봤을때는 비정상, )


4 Online Anomaly Detection (실시간 이상탐지)
    실시간 데이터 수집체계가 구축되어 있는 환경에서 탐지
    어떻게 빠르게 처리하고 이상을 탐지할지 설계하는 것이 매우 중요
    실시간 데이터 처리를 위해 Light-weight 알고리즘 사용됨
    

5 Distributed Anomaly Detection (분포 이상탐지)
    관측치의 정상분포에서 벗어나는 이상데이터를 탐지
    Point Anomaly Detection 과 곂치는 유형
```

```python
★★★이상탐지 종류(Label(정답) 유무에 따른) ★★★

현재 보유하고 있는 데이터 성격에 따라 이상탐지의 종류 및 해결 방법론이 결정된
1. Supervised Anomaly Detection (은 정답이 있는것임, 그러나 대부분 없음)
    모든관측치에 대해 이상(abnormal) 과 정상(normal) 라벨이 붙어 있는 경우 
    충분히 학습할 만한 이상 데이터가 확보되어 있는상황
    일반적인문제는 이진분류로 해결가능(이상종류가 다양하면 Multi Classification 문제로 해결가능)
    이상데이터확보가 가능하나 적은수 데이터 존재하면 -> 소수 클래스문제 , 불균형 문제 발생 상황고려 (SMOTE, Over/Under-sampling 으로 해결)
    성능 평가시 Accuracy 관점이 아닌 Recall 관점 (실제이상에서 이상을 맞출확률)  에서의 성능평가도 고려해야함 (이상데이터가 매우적어 ACC는 높게 나타날가능성큼)

장점:  양성/ 불량 판정 정확도가 높다(?) 트리계열의 알고리즘 적용(label이 있음으로 ) (Random forest, XGboost , LightGBM)
단점: Class -Imbalance 문제 (정상과 비정상이 불균형 하다)


2. Semi- Supervised Anomaly Detection
    이상에 대한 레벨이 없으며 , 정상 데이터만 보유하고 있는상황, 충분히 학습 할만한 이상테이터가 없을시 정상데이터만 사용해 이상탐지
    정상데이터의 패턴을 학습해 이상데이터가 들어올시 정상 데이터와 차이점을 파악하여 이상ㅌ데이터를 탐지
    정상데이터를 둘러싸고 있는 discriminative boundary(구분선) 를 설정하고 , 이 구분선을 최대한 좁혀 밖에 있는 데이터들을 이상으로 간주하는것
    대표적 방법은 One-classs SVM


3. Unsupervised Anomaly Detection
    정상과 이상에 대한 라벨이 모두 존재하지 않는경우
    모든 데이터를 활용하여 학습
    일반적으로 딥러닝알고리즘을 사용하여 정상패턴을 학습하고 ,학습된 모델을 통해 이상데이터를 탐짖함
    Autoendcoder를활용(정상데이터로 학습하고 , 잔차 (ex, MSE) 를 계산하여 Threshhold 적용

현업에서는 해결하기 힘듬
정상데이터를 정의하는 단계부터 시작(추정)
-현업 전문가와 협업 필요 

정상데이터를 추정하는단계 걸치면 Semi-supervised anomaly Detection 과 같은것 아닌가-? 
    -> 정상데이터를 추정하고 다시 알고리즘으로 추정하는 것임으로 같지않다( 즉, 라벨이 없는 상태에서 내가 이게 정답이라고 추정한상태이기때문에 같지않다. / 가정일 뿐 사실이 아니기때문)


장점: 라벨링 과정이 필요없음
단점: 양성/불량 판정 정확도가 높지 않고, Hyper Parameter 에 매우 민감함  (Dicision tree 보면 밑에 몇개까지 만들것인가 이런게 hyper parameter 임. )



일반적인 세상은 2, 3 번이 대부분임. 
```

```python
이상탐지 종류(Train Data에 따른)

새로운 관측치가 기존분포 (In-distribution)에 속하는지 , 기존분포를 벗어나는지 (Out-of-distribution) 를 구분
학습할 데이터를 어떻게 정의하는지에 따라 문제의 성격과 해결 방법론이 달라짐
1. Outlier Detection
    훈련데이터 셋에 정상샘플과 이상치 샘플을 모두 포함하고 있음. 



2. Novelty(참신 , 진귀) Detection
    학습할 데이터가 이상치에 의해 오염되지 않았다는 가정(이상치가 없음)
    새로운 패턴의 데이터를 찾는 것에 관심 (고양이종류 사진 맞추는 모델에 고양이 사진주고 고양이 아니다라고했을때 -> 새로운현상,아이디어 만들고자할떄) 
    훈련데이터가 특이치 데이터를 포함하지 않은 채 학습하고 예측시 새로운 관측치인지 확인한다
```

```python
이상탐지의 목적은 이상한 것을 사전에 빠르게 찾아야한다.

1 정상(normal) 데이터 정의
    도메인별 특별한 이상이 없는 정상적인 데이터 , 이상이라고 정의된 데이터 이외의 데이터
    제조도메인에서는 제품의 명확한 스펙이 존재하기 때문에 스펙안에 위치한 정상 제품의 데이터

2 이상(abnormal) 데이터 정의 
    사전에 감지하고 싶은 이상한 데이터
    제조 도메인에서 고장이 발생전 이상을 탐지하는 것이 주목적이기에 스펙의 상한치, 하한치에 근접한 제품이 이상데이터를 만들확률이 높음
    시계열 센서 뎅이터의 경우 오차범위를 지정하고 오차범위 진입시 이상으로 판단

3 데이터 수집 및 추출 과정 오류 확인
    데이터 정합성(신회성) 검증 : 수집 및 추출/변환 과정에서 오류가 없었는지 확인
    데이터 ETL 과정: 추출(Extract), 변환(Transform), 로드(Load) 각단계별 이슈 확인
        ex)센서 데이터 수집과정에서 센서 고장의로 -999 ,999 값수집.
        ex)데이터 변환 과정에서 Meta data(데이터 변환기준) 의 예상치 못한 변경으로 인해 raw data가 잘못 변환되는 경우
        ex)data 수집 server shutdown 및 점검으로 인한 down 시 수집 데이터의 공백 발생

4 Outlier Detection 방법 선정
        데이터 type/Label/Train DAta 에따라서 방법 선정

5 평가방법 선정
    scoring 평가 : 이상탐지 있으면 이상인지 아닌지 알려줘야함 -> 이걸 점수로 표현함. 
                    Threshold(임계치)의 순준에 따라 경고의 수준을 전달 ex)경광등
        ex) 0.5 이상이면 노랑 07. 이상이면 주황 , 0.9 이상이면 빨간

    Prediction 평가 : Threshold(임계치)를 설정하여 이상(1) 정상(0)을 구분하여 전달
        ex)아주 risky 한 공정에서는 이런 방법을 사용할 필요도 있음 

    민감도 설정 : 민감도 설정(threshold)에 따라 탐지하는 이상의 개수와 Point 가 달라짐


6 모니터링 및 대응 프로세스 수립
    이상을 감지하기 위한 현업 모니터링 시스템 설계
        Ex)대시보드, 메세지 알람, email 알람 등
    이상감지후 점검하기 위한 현업 Action 프로세스 설계
        Ex)유선확인 , 전문가 점검, 주요 지표 수치 확인 등
    false Alarm (오탐지) 이벤트에 대한 대응 체계 마련
    정상 데이터 패턴 변환에 대응하기 위한 재학습 체계수립 

```
